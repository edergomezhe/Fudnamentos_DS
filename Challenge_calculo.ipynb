f(x) = 2x^2 - 4x - 1

f'(x) = 4x - 4

f''(x) = 4

## Grafica f(x)=2x^2 -4x-1 en 3D.

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x):
    res = (2 * (x**2)) - (4 * x) - 1
    return (res)

resolucion = 100

X = np.linspace(-10, 10, resolucion)
Y = np.linspace(-10, 10, resolucion)

X, Y = np.meshgrid(X, Y)

Z = f(X) 

# Gráficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap=cm.cool, linewidth=0)

fig.colorbar(surf)
```
