```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

# Ejercicio 1: 

Demuestra que la matriz de correlación es una matriz simétrica por los siguientes métodos

*Usando Transpuesta

*Usando las matrices triangulares

```python
data_corr = data1.corr()
data_corr
```

```python
data_corr_t = data_corr.T
data_corr_t2 = data_corr_t.T
```

```python
data_corr == data_corr_t2
```

```python
data_corr_tri1 = np.triu(data_corr, 1)
data_corr_tri2 = np.tril(data_corr, 0)
data_corr_tri = data_corr_tri1 + data_corr_tri2
data_corr == data_corr_tri
```

# Ejercicio 2
Obten la matriz inversa de la matriz de correlación

```python
data_i = np.linalg.inv(data_corr)
data_i
```

# Ejercicio 3
¿Qué operación(es) harías para obtener la matriz identidad?

```python
data_ident = np.matmul(data_corr, data_i)
data_ident
```
